```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
csv_data = """
Month,Store,Sales
Jan,Store A,12000
Jan,Store B,15000
Feb,Store A,18000
Feb,Store B,17000
Mar,Store A,16000
Mar,Store B,19000
 """

with open('sales_data.csv','w')as file:
  file.write(csv_data)
```

```python
df = pd.read_csv('sales_data.csv')
print(df)
```

```python
# 2. Total Sales Per Store

total_sale = df.groupby('Store')['Sales'].sum()
print(total_sale)
higher_sale = total_sale.sort_values(ascending=False).iloc[0]
print(higher_sale)
```

```python
# 3. Monthly Sales Trend
sale_month = df.groupby('Month')['Sales'].sum()
print(sale_month)
```

```python
# 4. Chart: Monthly Sales by Store

store_a =df[df['Store']=='Store A']
store_b =df[df['Store']=='Store B']
plt.plot(store_a['Month'],store_a['Sales'], marker='o')
plt.plot(store_b['Month'],store_b['Sales'],marker='o')
#plt.plot(df['Month'],df['Store'],kind='line',marker='d')
plt.title('Monthly Sales by Store')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Bonus Bar chart of total sales per store

print(sale_month)
plt.bar(sale_month.index, sale_month.values,color='skyblue')
plt.title('Total sales per store')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.tight_layout()
plt.grid(True)
plt.show()
```
