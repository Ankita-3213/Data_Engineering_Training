```python
import pandas as pd

df = pd.read_csv('/content/students.csv')
#print(df)
print(df.head(3))
```

```python
# 2. Average scores

df['Average']=df[['Math','Science','English']].mean(axis=1) # axis =1 across row
print(df['Average'].max())
top= df[df['Average']==df['Average'].max()]
print(top)
```

```python
# 3.Filter Students
# 3.1 more than 80 in Science
sci_top= df[df['Science']>80] # using only df['Science']>80 gives true or false values
print('Students with more than 80 in science:\n',sci_top)
```

```python
# 3.2 all students with an average score below 70.
avg_below = df[df['Average']<70]
print('Students with average less than 70: \n',avg_below)
```

```python
# 4.Sort students by English score (descending) and print the top 3.
English_top = df.sort_values(by='English',ascending=False)
print(English_top.head(3))
```

```python
# 5.1 Create a bar chart showing studentsâ€™ names on the X-axis and their average
#scores on the Y-axis.
import matplotlib.pyplot as plt
df['Average']=df[['Math','Science','English']].mean(axis=1)

plt.figure(figsize=(8,5))
plt.bar(df['Name'],df['Average'],color='teal')
plt.xlabel('Students')
plt.ylabel('Average marks')
plt.xticks(rotation=30)
plt.grid(True)
plt.show()
```

```python
# 5.2 Create a line chart for all studentsâ€™ scores in Math, Science, and English
import matplotlib.pyplot as plt
df.set_index('Name' , inplace=False)
df[['Math','Science','English']].plot(kind='line', marker='d')

plt.title('Students scores')
plt.xlabel('Students')
plt.ylabel('Scores')
plt.grid(True)
plt.tight_layout()
```
