```python
import numpy as np
import pandas as pd
```

```python
# Reading data
customer = pd.read_csv('customers.csv')
customer
```

```python
order = pd.read_csv('orders.csv')
order
```

```python
order.info()
```

```python
# changing order_date and delivery_date to datetime
order['order_date'] = pd.to_datetime(order['order_date'],format='%d-%m-%Y', errors='coerce')
order['delivery_date'] = pd.to_datetime(order['delivery_date'],format='%d-%m-%Y', errors='coerce')
```

```python
order
```

```python
order.info()
```

```python
order.loc[order['item_name'] == 'Smartwatch', 'delivery_date'] = (order.loc[order['item_name']=='Smartwatch', 'order_date'] + pd.Timedelta(days=7))
```

```python
# Calculating delivery delays
order['delay_days'] = (order['delivery_date'] - order['order_date']).dt.days
order['delay'] = np.where(order['delay_days'] > 6, 1, 0)
```

```python
order
```

```python
merged = pd.merge(order, customer[['customer_id', 'customer_name']], on='customer_id', how='left')
merged
```

```python
# Top delays
top_delayed = merged.groupby('customer_name')['delay'].sum().sort_values(ascending=False)
print("Top delayed customers:")
print(top_delayed)
```

```python
order.to_csv('cleaned_order.csv', index=False)
```
