```python
import numpy as np
import pandas as pd
import datetime as dt
```

```python
customer = pd.read_csv('customers.csv')
customer
```

```python
order=pd.read_csv('orders.csv')
order
```

```python
order.info()
```

```python
# Converting order_date and delivery_date from object to datetime
order['order_date'] = pd.to_datetime(order['order_date'], format="%d-%m-%Y", errors='coerce')
order['delivery_date'] = pd.to_datetime(order['delivery_date'], format="%d-%m-%Y", errors='coerce')
```

```python
order.info()
```

```python
# Days delayed
order['days_delayed'] = (order['delivery_date'] - order['order_date']).dt.days
order
```

```python
# Handling missing values
order['item_name'] = order['item_name'].fillna("Unknown")
```

```python
order.loc[(order['item_name']=='Acting Guides') & (order['customer_id'].isna()),'customer_id']=106

```

```python
order
```

```python
# Filling dates and delay

order.loc[order['item_name'] == 'Astronomy Telescope', 'order_date'] = '2025-05-08'

order['delivery_date'] = order['order_date'] + pd.to_timedelta(4, unit='D')
order['days_delayed'] = (order['delivery_date'] - order['order_date']).dt.days
```

```python
order
```

```python
# Filling missing values for customers

customer['name'] = customer['name'].fillna('Unknown')
customer['email'] = customer['email'].fillna('Not Provided')
customer['location'] = customer['location'].fillna('Not Provided')
```

```python
customer
```

```python
# Merging orders with customers
merged = order.merge( customer[['customer_id', 'name', 'location']], on='customer_id', how='left')
```

```python
merged
```

```python
# The order with no location will be problematic to deliver it
problematic_order = merged[merged['location'] == 'Not Provided']
```

```python
problematic_order
```

```python
# Saving clean data
customer.to_csv("cleaned_customer.csv", index=False)
order.to_csv("cleaned_order.csv", index=False)
merged.to_csv("clean_merged.csv", index=False)
```
